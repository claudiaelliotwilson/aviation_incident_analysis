let's start with the incident_date column. there's another column called date that has the same information but with a day of week even. Let's drop the incident date column and just use the date column.

```python
df_clean = df_clean.drop(columns = ['Incident_Date'])
```

```python
# okay now let's fix the date column

df_clean[['Day_of_Week','Date']] = df['Date'].str.split(" ", n=1, expand=True)
df_clean['Date'] = pd.to_datetime(df_clean['Date'])
```

```python
df_clean.head()
```

```python
df_clean['Year'] = df_clean['Date'].dt.year
df_clean['Month'] = df_clean['Date'].dt.month_name()
df_clean['Day'] = df_clean['Date'].dt.day
df_clean['Day_of_Week'] = df_clean['Date'].dt.day_name()
```

```python
df_clean.head()
```

```python
#now let's do incident category
# we want to split by the pipe

df_clean[['Incident_Gen_Category','Incident_Detail_Category']] = df['Incident_Category'].str.split('|', n=1, expand=True)
df_clean = df_clean.drop(columns = ['Incident_Category'])
df_clean.head()
```
